```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')


# Maxpulse histogram
plt.figure(figsize=(10, 5))
plt.hist(df['Maxpulse'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Maxpulse')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Calories histogram
plt.figure(figsize=(10, 5))
plt.hist(df['Calories'].dropna(), bins=20, color='salmon', edgecolor='black')
plt.title('Histogram of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Maxpulse
plt.subplot(1, 2, 1)
plt.hist(df['Maxpulse'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Maxpulse Histogram')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')
plt.grid(True)

# Calories
plt.subplot(1, 2, 2)
plt.hist(df['Calories'].dropna(), bins=20, color='salmon', edgecolor='black')
plt.title('Calories Histogram')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```
